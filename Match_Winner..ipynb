```python
import pandas as pd
df = pd.read_csv("Match Winner.csv")

df.columns = df.columns.str.strip()

print(" Dataset Preview:")
display(df.head())

print("\n Dataset Info:")
df.info()

print("\n Missing Values in each column:")
display(df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nDuplicate rows found: {duplicates}")
df = df.drop_duplicates()
print("Duplicate rows removed.")

print("\n Summary Statistics:")
display(df.describe())

print("\n Match Results Count (FTR column):")
display(df['FTR'].value_counts())

print("\n Wins per Team (Home and Away):")

home_wins = df[df['FTR']=='H']['HomeTeam'].value_counts()

away_wins = df[df['FTR']=='A']['AwayTeam'].value_counts()

total_wins = home_wins.add(away_wins, fill_value=0)
display(total_wins.sort_values(ascending=False))

df.to_csv("Match_Winner_Cleaned.csv", index=False)
print("\n Cleaned dataset saved as 'Match_Winner_Cleaned.csv'")
```
