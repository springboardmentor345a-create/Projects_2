```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(" Libraries imported successfully.")
```

```python
file_name = "Goals & Assist.xlsx"  
df = pd.read_excel(file_name)

print(" Dataset loaded successfully")
```

```python
print(" Dataset Preview:")
display(df.head())
```

```python
print("\n Dataset Info:")
df.info()
```

```python
print("\n Missing Values in each column:")
display(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows found: {duplicates}")
df = df.drop_duplicates()
print(" Duplicate rows removed")
```

```python
print("\n Summary Statistics:")
display(df.describe())
```

```python
player_col = "Player"
goals_col = "Goals"
assists_col = "Assists"

required = [player_col, goals_col, assists_col,]
missing = [c for c in required if c not in df.columns]

if missing:
    print(f" Missing columns: {missing}")
    print("Columns found in file:")
    print(list(df.columns))
else:
    print("All required columns exist")
```

```python
top_goals = df.sort_values(by=goals_col, ascending=False)[[player_col, goals_col]]
print("\n Top 10 Goal Scorers:")
display(top_goals.head(10))
```

```python
top_assists = df.sort_values(by=assists_col, ascending=False)[[player_col, assists_col]]
print("\n Top 10 Assist Providers:")
display(top_assists.head(10))
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=player_col, y=goals_col, data=top_goals.head(10))
plt.xticks(rotation=45)
plt.title("Top 10 Goal Scorers")
plt.xlabel("Player")
plt.ylabel("Goals")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=player_col, y=assists_col, data=top_assists.head(10))
plt.xticks(rotation=45)
plt.title("Top 10 Assist Providers")
plt.xlabel("Player")
plt.ylabel("Assists")
plt.tight_layout()
plt.show()
```
