```python
import pandas as pd
path = "/content/drive/MyDrive/Colab Notebooks/Infosys Springboard/Data Set + Python files/Goals & Assist.xlsx"
goals_assist = pd.read_excel(path, engine='openpyxl')
print(goals_assist.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numeric_df_match_winner = goals_assist.select_dtypes(include='number')
correlation_matrix = numeric_df_match_winner.corr()
plt.figure(figsize=(20, 12))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',linewidths=2)
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```

Analyzing the correlation matrix for 'Goals + Assists':

```python
target_correlation = correlation_matrix['Goals + Assists'].sort_values(ascending=False)
print("Correlation with 'Goals + Assists':\n", target_correlation)

# Define a threshold for correlation (e.g., absolute value > 0.2)
correlation_threshold = 0.2

# Identify columns to keep (including the target itself)
columns_to_keep = target_correlation[abs(target_correlation) > correlation_threshold].index.tolist()

# Identify columns to drop
all_columns = goals_assist.select_dtypes(include='number').columns.tolist()
columns_to_drop = [col for col in all_columns if col not in columns_to_keep]

print("\nColumns to consider keeping (absolute correlation > {}):\n".format(correlation_threshold), columns_to_keep)
print("\nColumns to consider dropping (absolute correlation <= {}):\n".format(correlation_threshold), columns_to_drop)
```

```python
columns_to_drop = ['90s Played', 'Goals + Assists', 'Goals Per 90', 'Assists Per 90', 'Goals + Assists Per 90', 'Non-Penalty Goals Per 90', 'Non-Penalty Goals + Assists Per 90', 'xG Per 90', 'xAG Per 90', 'xG + xAG Per 90', 'npxG Per 90', 'npxG + xAG Per 90', 'Yellow Cards', 'Red Cards', 'npxG + xAG']
goals_assist = goals_assist.drop(columns=columns_to_drop)
goals_assist.head()
```
