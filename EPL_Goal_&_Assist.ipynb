Analyzing the correlation matrix for 'Goals + Assists':

```python
target_correlation = correlation_matrix['Goals + Assists'].sort_values(ascending=False)
print("Correlation with 'Goals + Assists':\n", target_correlation)

# Define a threshold for correlation (e.g., absolute value > 0.2)
correlation_threshold = 0.2

# Identify columns to keep (including the target itself)
columns_to_keep = target_correlation[abs(target_correlation) > correlation_threshold].index.tolist()

# Identify columns to drop
all_columns = goals_assist.select_dtypes(include='number').columns.tolist()
columns_to_drop = [col for col in all_columns if col not in columns_to_keep]

print("\nColumns to consider keeping (absolute correlation > {}):\n".format(correlation_threshold), columns_to_keep)
print("\nColumns to consider dropping (absolute correlation <= {}):\n".format(correlation_threshold), columns_to_drop)
```

```python
# Checking for outliers
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named 'goals_assist'
# Select all columns that contain numeric data types (int, float, etc.)
numerical_cols = goals_assist.select_dtypes(include=['number']).columns.tolist()

print(f"Generating box plots for columns: {numerical_cols}")

# Loop through the list of numerical columns
for column in numerical_cols:
    # Create a new plot figure for each column
    plt.figure(figsize=(8, 5))

    # Generate the box plot using the specific column from the goals_assist DataFrame
    sns.boxplot(x=goals_assist[column])

    # Set the title and label
    plt.title(f"Box Plot for Outlier Detection in '{column}'")
    plt.xlabel(column)

    # Display the plot
    plt.show()
```

```python
# Removing outliers
import pandas as pd
import numpy as np

def remove_outliers_simple(df):

    numerical_cols = df.select_dtypes(include=['number']).columns.tolist()


    df_filtered = df.copy()

    for column in numerical_cols:
        Q1 = df_filtered[column].quantile(0.25)
        Q3 = df_filtered[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR


        df_filtered = df_filtered[(df_filtered[column] >= lower_bound) & (df_filtered[column] <= upper_bound)]

    rows_removed = len(df) - len(df_filtered)
    print(f"Original shape: {df.shape}")
    print(f"Cleaned shape: {df_filtered.shape}")
    print(f"Total rows removed as outliers across all numerical columns: {rows_removed}")

    return df_filtered

# Apply the function to your DataFrame
match_winner_cleaned = remove_outliers_simple(goals_assist)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Select only the numerical columns from the CLEANED DataFrame
numerical_cols_cleaned = match_winner_cleaned.select_dtypes(include=['number']).columns.tolist()

print(f"Generating NEW box plots for columns in cleaned data: {numerical_cols_cleaned}")

for column in numerical_cols_cleaned:
    plt.figure(figsize=(8, 5))
    # Plot using the CLEANED DataFrame
    sns.boxplot(x=match_winner_cleaned[column])
    plt.title(f"Cleaned Box Plot for '{column}' (Outliers Removed)")
    plt.xlabel(column)
    plt.show()
```
