```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(" Libraries imported successfully")
```

```python
file_name = "ScoreSight_ML_Season_LeagueWinner_Champion (1).csv"
df = pd.read_csv(file_name)

print(" Dataset loaded successfully")
```

```python
print(" Columns in your dataset:")
display(df.columns)
```

```python
df.columns = df.columns.str.strip()  
print(" Dataset Preview:")
display(df.head())
```

```python
print("\n Dataset Info:")
display(df.info())

print("\n Missing values in each column:")
display(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows found: {duplicates}")
df = df.drop_duplicates()
print(" Duplicate rows removed")
```

```python
print("\n Summary Statistics:")
display(df.describe())
```

```python
player_col = "Player"  # replace if your CSV has different name
goals_col = "Goals"    # replace if different
assists_col = "Assists"  # replace if different

# Check if 'Team' exists
team_col = None
if "Team" in df.columns:
    team_col = "Team"
elif "Club" in df.columns:
    team_col = "Club"

required = [player_col, goals_col, assists_col]
missing = [c for c in required if c not in df.columns]
if missing:
    print(f" Missing required columns: {missing}")
else:
    print(" All required columns are present")
if team_col:
    print(f" Team/Club column detected: {team_col}")
else:
    print("ℹ No Team/Club column found, will skip team-related analysis")
```

```python
if goals_col in df.columns:
    columns_to_show = [player_col, goals_col]
    if team_col in df.columns:
        columns_to_show.insert(1, team_col)

    top_goals = df.sort_values(by=goals_col, ascending=False)[columns_to_show].head(10)
    print("\n Top 10 Goal Scorers:")
    print(top_goals)
else:
    print("'Goals' column not found in dataset")
```

```python
if assists_col in df.columns:
    columns_to_show = [player_col, assists_col]
    if team_col in df.columns:
        columns_to_show.insert(1, team_col)

    top_assists = df.sort_values(by=assists_col, ascending=False)[columns_to_show].head(10)
    print("\n Top 10 Assist Providers:")
    print(top_assists)
else:
    print(" 'Assists' column not found in dataset")
```
